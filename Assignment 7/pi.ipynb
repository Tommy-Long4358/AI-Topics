```python
import random
import matplotlib.pyplot as plt
import math

'''
- Program should generate n random points of (x,y) and determine if its below the "curve"
- n has to be 10^3, 10^4, 10^5, or 10^6
- Compare each approximated pi with the actual math.pi amount and determine its error rate
'''

# Error rate to determine how close the approximation is to the real value.
def errorRate(piPrime):
    return "{0:.4f}".format(abs(((math.pi - piPrime) / math.pi) * 100))

for n in range(3, 7):
    # List of points under the curve
    pointsUnderA = []

    # List of points above the curve
    pointsAboveA = []

    # Total number of points in the square which is nPointsList[n]
    square = 10**n

    for i in range(square):
        # generate n random points of (x, y) such that 0 < x < 1 and 0 < y < 1
        [x, y] = (random.random(), random.random())

        # Add to certain set if distance from origin (0, 0) is less than 1.
        if math.sqrt(x**2 + y**2) < 1:
            pointsUnderA.append([x, y])
        else:
            pointsAboveA.append([x, y])
    
    # Calculate approximation of pi based on set of 10^n random coordinates under the curve.
    piApprox = (len(pointsUnderA) / square) * 4

    # Output format
    print("n = 10 ^", n, "pi = ", "{0:.6f}".format(piApprox), "error = ", errorRate(piApprox), "%")

    if 10**n == 10**4:
        # Plot points
        plt.title("Estimating  using simulation of 10**4")
        zip_pointsUnderA = zip(*pointsUnderA)
        plt.scatter(*zip_pointsUnderA, marker='.', color='b', s = 1)

        zip_pointsAboveA = zip(*pointsAboveA)
        plt.scatter(*zip_pointsAboveA, marker='.', color='r', s = 1)
        plt
```
