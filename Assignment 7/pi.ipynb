```python
import random
import matplotlib.pyplot as plt
import math

'''
- Program should generate n random points of (x,y) and determine if its below the "curve"
- n has to be 10^3, 10^4, 10^5, or 10^6
- Compare each approximated pi with the actual math.pi amount and determine its error rate
'''

def errorRate(pie):
    return "{0:.4f}".format(abs(((math.pi - pie) / math.pi) * 100))

nPointsList = [10**3, 10**4, 10**5, 10**6]
for n in range(len(nPointsList)):
    pointsUnderA = set()
    pointsAboveA = set()

    for i in range(nPointsList[n]):
        # generate n random points of (x, y)
        (x, y) = (random.random(), random.random())

        # Add to certain set if distance from origin (0, 0) is less than 1
        if math.sqrt(x**2 + y**2) < 1:
            pointsUnderA.add((x, y))
        else:
            pointsAboveA.add((x, y))
    
    # Calculate pi approximation based on set
    square = nPointsList[n]
    piEstimate = (len(pointsUnderA) / square) * 4

    # Output format
    print("n = 10 ^", 3 + n, "pi = ", "{0:.6f}".format(piEstimate), "error = ", errorRate(piEstimate), "%")

    if nPointsList[n] == 10**4:
        plt.title("Estimating pi using simulation of 10**4")
        zip_pointsUnderA = zip(*pointsUnderA)
        plt.scatter(*zip_pointsUnderA, marker='.', color='b')

        zip_pointsAboveA = zip(*pointsAboveA)
        plt.scatter(*zip_pointsAboveA, marker='.', color='r')

        plt
```
